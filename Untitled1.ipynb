```python
# 曲名のリスト（ここにChatGPTが提案した曲を入れていく）
track_names = [
    "Weightless - Marconi Union",
    "Night Owl - Galimatias",
    "Elegy - Lisa Gerrard",
    "Calm Down - Ambient Forest",
    "Departure - Hammock"
]

track_uris = []

# 各曲名でSpotify検索 → 最初の候補のURIを取得
for name in track_names:
    results = sp.search(q=name, limit=1, type='track')
    items = results['tracks']['items']
    if items:
        track_uris.append(items[0]['uri'])
        print(f"追加準備OK: {name}")
    else:
        print(f"見つからなかった: {name}")

# 曲をプレイリストに追加
sp.playlist_add_items(playlist_id=playlist['id'], items=track_uris)

print("プレイリストに曲を追加しました！")
```

```python
# ChatGPTが出してくれた曲リストをここに貼る（曲名 - アーティスト名）
gpt_tracks = """
Dreams - Nuages
Inside - Kiasmos
Reverie - Nils Frahm
Dawn Chorus - Hidden Orchestra
Night Rain - Oatmello
Low - Haring
Argo - Goldmund
Signal - Tycho
Opening - Philip Glass
2AM - Yoe Mase (Instrumental ver.)
"""  # ← ここに50曲でも100曲でもOK！

# 行ごとに分割して検索＆追加
track_names = [line.strip() for line in gpt_tracks.strip().split('\n') if line.strip()]
track_uris = []

for name in track_names:
    results = sp.search(q=name, limit=1, type='track')
    items = results['tracks']['items']
    if items:
        track_uris.append(items[0]['uri'])
        print(f"追加準備OK: {name}")
    else:
        print(f"見つからなかった: {name}")

# 曲をプレイリストに追加（最大100曲まで対応）
sp.playlist_add_items(playlist_id=playlist['id'], items=track_uris)
print("ChatGPTのリストから曲をプレイリストに追加しました！")
```
